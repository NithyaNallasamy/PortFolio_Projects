```python
from bs4 import BeautifulSoup
import requests,openpyxl
import re
import pandas as pd
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
```

```python
# Downloading imdb top 250 movie's data


excel = openpyxl.Workbook()
print(excel.sheetnames)
sheet = excel.active
sheet.title = "Top Rated Movies"
print(excel.sheetnames)
sheet.append(['Movie Rank','Movie Name','Year of Release','IMDB Rating'])

try: 
    head = {
          'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36',
          'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
          'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
          'Accept-Encoding': 'none',
          'Accept-Language': 'en-US,en;q=0.8',
          'Connection': 'keep-alive',
          'refere': 'https://example.com',
          'cookie': """your cookie value ( you can get that from your web page) """
        }
    url = 'http://www.imdb.com/chart/top'
    
    

    session = requests.Session()
    retry = Retry(connect=3, backoff_factor=0.5)
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)

    session.get(url)

    response = requests.get(url,headers=head)
    soup = BeautifulSoup(response.text, "html.parser")
    movies = soup.find('ul',class_="ipc-metadata-list ipc-metadata-list--dividers-between sc-3a353071-0 wTPeg compact-list-view ipc-metadata-list--base").find_all('li')
    
    for movie in movies:
        
        name = movie.find('div',class_="sc-14dd939d-0 fBusXE cli-children").a.text.split('.')[1]
        rank = movie.find('div',class_="sc-14dd939d-0 fBusXE cli-children").a.text.split('.')[0]
        year = movie.find('div',class_="sc-14dd939d-0 fBusXE cli-children").span.text
        rating = movie.find('span',class_="sc-14dd939d-1 PnacM").div.span.text
        print(rank,name,year,rating)
        sheet.append([rank,name,year,rating])
    
except Exception as e:

    print(e)
    
    
excel.save('IMDB Movie Rating.xlsx')
```
